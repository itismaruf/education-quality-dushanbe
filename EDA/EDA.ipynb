## Группировки для выявления основных лидеров и зон риска


```python
# 6. Комбинации район + тип школы: где лучшие результаты
grouped_rt = (
    df.groupby(["Район", "Тип_школы"])
      .agg(
          Ученики=("ID_ученика", "count"),
          Ср_балл=("Средний_балл", "mean"),
          Ср_экзамен=("Рез_экзамена", "mean"),
          Ср_посещаемость=("Посещаемость_%", "mean"),
          Ср_стресс=("Уровень_стресса_перед_экзаменом", "mean")
      )
      .round(2)
      .sort_values("Ср_экзамен", ascending=False)
)
grouped_rt
```

```python
# 7. Рейтинг районов по интегральной эффективности
district_rank = (
    df.groupby("Район")
      .agg(
          Ученики=("ID_ученика", "count"),
          Ср_экзамен=("Рез_экзамена", "mean"),
          Ср_балл=("Средний_балл", "mean"),
          Ср_посещаемость=("Посещаемость_%", "mean"),
          Индекс_качества=("Индекс_качества_школы", "mean")
      )
)
district_rank["Эффективность"] = (
    0.5 * district_rank["Ср_экзамен"]
    + 0.3 * district_rank["Индекс_качества"]
    + 0.2 * district_rank["Ср_посещаемость"]
).round(2)
district_rank = district_rank.round(2).sort_values("Эффективность", ascending=False)
district_rank
```

```python
# 8. Группировка по уровню доступа к ресурсам (квартили)
df["Квартиль_ресурсов"] = pd.qcut(
    df["Доступ_к_ресурсам"],
    q=4,
    labels=["Q1: низкий", "Q2", "Q3", "Q4: высокий"]
)
resource_groups = (
    df.groupby("Квартиль_ресурсов", observed=False)
      .agg(
          Ученики=("ID_ученика", "count"),
          Ср_экзамен=("Рез_экзамена", "mean"),
          Ср_балл=("Средний_балл", "mean"),
          Ср_стресс=("Уровень_стресса_перед_экзаменом", "mean"),
          Ср_посещаемость=("Посещаемость_%", "mean")
      )
      .round(2)
)
resource_groups
```

```python
# 14. Распределения + взаимосвязи + кандидаты для feature engineering
from math import ceil
from plotly.subplots import make_subplots

TARGET = "Рез_экзамена"
ID_COL = "ID_ученика"

df_eda = df.copy()

numeric_cols = df_eda.select_dtypes(include=[np.number]).columns.tolist()
if ID_COL in numeric_cols:
    numeric_cols.remove(ID_COL)

feature_num_cols = [c for c in numeric_cols if c != TARGET]
cat_cols = df_eda.select_dtypes(include=["object", "category"]).columns.tolist()

print(f"Rows: {df_eda.shape[0]}, Cols: {df_eda.shape[1]}")
print(f"Numeric features: {len(feature_num_cols)}, Categorical features: {len(cat_cols)}")
```

```python
# 14.1 Цифровая сводка распределений
display(df_eda[feature_num_cols].describe(percentiles=[0.05, 0.25, 0.5, 0.75, 0.95]).T.round(2))
```

```python
# 14.2 Визуально: гистограммы всех числовых признаков
ncols = 3
nrows = ceil(len(feature_num_cols) / ncols)
fig_dist = make_subplots(rows=nrows, cols=ncols, subplot_titles=feature_num_cols)

for i, col in enumerate(feature_num_cols, start=1):
    r = (i - 1) // ncols + 1
    c = (i - 1) % ncols + 1
    fig_dist.add_trace(
        go.Histogram(
            x=df_eda[col],
            nbinsx=35,
            marker=dict(color="#1f77b4"),
            showlegend=False,
            hovertemplate=f"{col}: %{{x}}<br>Count: %{{y}}<extra></extra>"
        ),
        row=r,
        col=c,
    )

fig_dist.update_layout(
    title="Распределения числовых признаков",
    height=320 * nrows,
    bargap=0.08,
)
fig_dist.show()
```

```python
# 14.3 Корреляции (Spearman): числовые признаки + таргет
corr_s = df_eda[feature_num_cols + [TARGET]].corr(method="spearman")
fig_corr = px.imshow(
    corr_s.round(2),
    text_auto=True,
    color_continuous_scale="RdBu_r",
    zmin=-1,
    zmax=1,
    aspect="auto",
    title="Spearman-корреляции: числовые признаки и таргет",
)
fig_corr.show()
```

```python
# 14.4 Цифровой рейтинг связи признаков с таргетом
target_corr = corr_s[TARGET].drop(TARGET).sort_values(key=lambda s: s.abs(), ascending=False)
display(target_corr.to_frame("spearman_vs_target").round(3))

fig_target_corr = px.bar(
    target_corr.sort_values(),
    orientation="h",
    title="Сила связи числовых признаков с таргетом (Spearman)",
    labels={"value": "Spearman", "index": "Признак"},
)
fig_target_corr.show()
```

```python
# 14.5 Категориальные признаки vs таргет (и визуально, и цифрами)
cat_summary = []
for col in cat_cols:
    stats = (
        df_eda.groupby(col, dropna=False)[TARGET]
        .agg(target_mean="mean", count="count", target_std="std")
        .sort_values("target_mean", ascending=False)
    )
    spread = stats["target_mean"].max() - stats["target_mean"].min()
    cat_summary.append(
        {
            "feature": col,
            "levels": int(stats.shape[0]),
            "target_mean_spread": float(spread),
        }
    )

    display(stats.round(3))
    fig_box = px.box(
        df_eda,
        x=col,
        y=TARGET,
        color=col,
        points=False,
        title=f"{TARGET} по категориям: {col}",
    )
    fig_box.show()

cat_summary_df = pd.DataFrame(cat_summary).sort_values("target_mean_spread", ascending=False)
display(cat_summary_df.round(3))
```

```python
# 14.6 Финальные новые фичи (как в CatBoost) + численная проверка связи с таргетом
fe = df_eda.copy()

# Финальный набор новых признаков
fe["Бал_на_стресс"] = fe["Средний_балл"] / (fe["Уровень_стресса_перед_экзаменом"] + 1)
fe["Бал_на_уверенность"] = fe["Средний_балл"] * fe["Уверенность_в_себе"]
fe["Балл_минус_стресс"] = fe["Средний_балл"] - fe["Уровень_стресса_перед_экзаменом"]
fe["Пропуски_на_стресс"] = fe["Пропущенные_дни"] * fe["Уровень_стресса_перед_экзаменом"]
fe["Среда_школы_среднее"] = (
    fe["Индекс_качества_школы"] + fe["Доступ_к_ресурсам"] + fe["Образовательная_среда"]
) / 3
fe["Качество_на_среду"] = fe["Индекс_качества_школы"] * fe["Образовательная_среда"]
fe["Качество_на_ресурсы"] = fe["Индекс_качества_школы"] * fe["Доступ_к_ресурсам"]

candidate_cols = [
    "Бал_на_стресс",
    "Бал_на_уверенность",
    "Балл_минус_стресс",
    "Пропуски_на_стресс",
    "Среда_школы_среднее",
    "Качество_на_среду",
    "Качество_на_ресурсы",
]

# Цифровая проверка связи с таргетом
cand_corr = (
    fe[candidate_cols + [TARGET]]
    .corr(method="spearman")[TARGET]
    .drop(TARGET)
    .sort_values(key=lambda s: s.abs(), ascending=False)
)

display(cand_corr.to_frame("spearman_vs_target").round(3))

fig_cand = px.bar(
    cand_corr.sort_values(),
    orientation="h",
    title="Финальные новые фичи: связь с таргетом (Spearman)",
    labels={"value": "Spearman", "index": "Новая фича"},
)
fig_cand.show()

# Визуально: распределения финальных новых фич
fig_cand_dist = make_subplots(rows=3, cols=3, subplot_titles=candidate_cols)
for i, col in enumerate(candidate_cols, start=1):
    r = (i - 1) // 3 + 1
    c = (i - 1) % 3 + 1
    fig_cand_dist.add_trace(
        go.Histogram(x=fe[col], nbinsx=35, showlegend=False),
        row=r, col=c
    )

fig_cand_dist.update_layout(
    title="Распределения финальных новых фич",
    height=900,
    bargap=0.08,
)
fig_cand_dist.show()
```

```python
# 14.7 Бинаризация финальных фич (медианный порог) с оценкой прироста таргета
binary_source_cols = [
    "Бал_на_стресс",
    "Бал_на_уверенность",
    "Балл_минус_стресс",
    "Пропуски_на_стресс",
    "Среда_школы_среднее",
    "Качество_на_среду",
    "Качество_на_ресурсы",
]

rows = []
for col in binary_source_cols:
    s = fe[col].astype(float)
    thr = s.median()
    b = (s > thr).astype(int)
    means = fe.groupby(b)[TARGET].mean()
    lift = means.get(1, np.nan) - means.get(0, np.nan)
    rows.append(
        {
            "feature": col,
            "threshold_median": float(thr),
            "target_mean_if_1": float(means.get(1, np.nan)),
            "target_mean_if_0": float(means.get(0, np.nan)),
            "lift_1_minus_0": float(lift),
            "abs_lift": float(abs(lift)),
        }
    )

binary_eval = pd.DataFrame(rows).sort_values("abs_lift", ascending=False)
display(binary_eval[["feature", "threshold_median", "target_mean_if_1", "target_mean_if_0", "lift_1_minus_0"]].round(3))

fig_bin = px.bar(
    binary_eval.sort_values("lift_1_minus_0"),
    x="lift_1_minus_0",
    y="feature",
    orientation="h",
    title="Финальные новые фичи (бинарно): разница среднего таргета (1 - 0)",
    labels={"lift_1_minus_0": "Разница среднего таргета", "feature": "Фича"},
)
fig_bin.show()

print("Проверка финальных новых фич завершена: используем этот набор в CatBoost.")
```

## Итоговый вывод по финальным новым фичам (синхронизировано с CatBoost)

После анализа распределений, корреляций и пороговых проверок в EDA, финальный набор новых признаков для моделирования:

1. `Бал_на_стресс = Средний_балл / (Уровень_стресса_перед_экзаменом + 1)`
Причина: отражает академическую устойчивость под стрессом и стабильно связан с таргетом.

2. `Бал_на_уверенность = Средний_балл * Уверенность_в_себе`
Причина: объединяет успеваемость и психологическую готовность, усиливая разделение сильных и слабых профилей.

3. `Балл_минус_стресс = Средний_балл - Уровень_стресса_перед_экзаменом`
Причина: интерпретируемая шкала "успеваемость против давления".

4. `Пропуски_на_стресс = Пропущенные_дни * Уровень_стресса_перед_экзаменом`
Причина: фиксирует совместный риск двух негативных факторов.

5. `Среда_школы_среднее = (Индекс_качества_школы + Доступ_к_ресурсам + Образовательная_среда) / 3`
Причина: интегральный индекс условий обучения, менее шумный, чем отдельные компоненты.

6. `Качество_на_среду = Индекс_качества_школы * Образовательная_среда`
Причина: учитывает синергию инфраструктуры и образовательного климата.

7. `Качество_на_ресурсы = Индекс_качества_школы * Доступ_к_ресурсам`
Причина: реалистичный interaction-фактор качества школы и обеспеченности ресурсами.

### Практический итог

1. В EDA и в CatBoost используем одинаковый финальный набор из 7 новых фич.
2. Бинарные версии этих же фич оцениваем по `lift`, чтобы при необходимости добавить простые пороговые индикаторы.
3. Окончательное решение о включении признаков принимаем по приросту метрик на валидации.
